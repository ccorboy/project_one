```python
cleveland_df = pd.read_csv('processed.cleveland.csv', header = None)
cleveland_df
```

```python
# adding column headers to data
cleveland_df.columns = ["age", "sex", "cp", "trestbps", 
                        "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal","healthy"]
cleveland_df
```

```python
# replaced ? with 99999 
clean_cleveland_df = cleveland_df.replace("?", 99999)
clean_cleveland_df
```

```python
# reset index to age
clean_cleveland_df.set_index("age", inplace = True)
clean_cleveland_df
```

```python
# Saving cleaned data to use for ml
clean_cleveland_df.to_csv('cleaned_cleveland.csv')
```
